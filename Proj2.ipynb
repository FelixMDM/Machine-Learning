```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_file = 'CS170_Spring_2024_Small_data__41.txt'  # Replace with your actual file name
data = pd.read_csv(data_file, delim_whitespace=True, header=None)

data_file_large = 'CS170_Spring_2024_Large_data__41.txt'
data_large = pd.read_csv(data_file_large, delim_whitespace=True, header=None)

print(data.head())

class_labels = data.iloc[:, 0]
features = data.iloc[:, 1:]

class_labels_large = data_large.iloc[:, 0]
features_large = data_large.iloc[:, 1:]

print(class_labels.head())
print(features.head())

# Scatter plot of the first two features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=features.iloc[:, 5], y=features.iloc[:, 6], hue=class_labels, palette='viridis')
plt.xlabel('Feature 5')
plt.ylabel('Feature 6')
plt.title('Scatter Plot - Small Dataset')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x=features_large.iloc[:, 18], y=features_large.iloc[:, 35], hue=class_labels_large, palette='viridis')
plt.xlabel('Feature 18')
plt.ylabel('Feature 35')
plt.title('Scatter Plot - Large Dataset')
plt.show()
```
